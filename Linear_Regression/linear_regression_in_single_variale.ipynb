```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("C:\\Users\\parth\\Downloads\\homeprices.csv")
df
```

```python
%matplotlib inline
plt.ylabel("price(US$)---->")
plt.xlabel("area(sq.ft)---->")
plt.scatter(df.area,df.price,color="red",marker="x")
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[["area"]],df["price"])
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
y = reg.coef_*3300+reg.intercept_
y
```
