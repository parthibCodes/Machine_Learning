```python
import pandas as pd
from sklearn import linear_model
```

```python
df = pd.DataFrame({
    "Years of Experience":[2,5,1,3,6,4],
    "Funding(₹ Cr)":[1.5,3.0,0.8,2.2,5.0,3.5],
    "City":["Bangalore","Delhi","Mumbai","Kolkata","Bangalore","Delhi"],
    "Founder Background":["Engineer","MBA","Engineer","Dropout","MBA","Dropout"],
    "Success (%)":[65,80,50,60,90,75]
})
df.to_csv("C:\\Users\\parth\\OneDrive\\Desktop\\ML_CSV\\startupSuccess.csv",index=False)
```

```python
df = pd.read_csv("C:\\Users\\parth\\OneDrive\\Desktop\\ML_CSV\\startupSuccess.csv")
df
```

```python
dummies_city = pd.get_dummies(df["City"])
dummies_city
```

```python
dummies_background = pd.get_dummies(df["Founder Background"])
dummies_background
```

```python
merged = pd.concat([df,dummies_city,dummies_background],axis="columns")
merged
```

```python
final = merged.drop(["City","Founder Background","Kolkata","Engineer"],axis="columns")
final
```

```python
model = linear_model.LinearRegression()
```

```python
x = final.drop("Success (%)",axis="columns")
x
```

```python
y = df["Success (%)"]
y
```

```python
model.fit(x,y)
```

```python
model.score(x,y)
```

```python
model.predict([[3,2.5,0,0,1,0,0]])
```
