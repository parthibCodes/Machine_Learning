```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("C:\\Users\\parth\\OneDrive\\Desktop\\ML_CSV\\employee_attrition.csv")
df.head()
```

```python
x = df.drop('LeftCompany',axis="columns")
y = df.LeftCompany
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train,y_train)
```

```python
features_importances = model.feature_importances_
feature_names = df.columns[:-1]
```

```python
forest_importances = pd.Series(features_importances,feature_names)
fig,ax = plt.subplots(figsize=(10,6))
forest_importances.plot.bar(ax=ax)
ax.set_title("Feature Importances")
ax.set_ylabel("Importance")
ax.set_xlabel("feature_names")
# plt.xticks()
plt.show()
```

```python
model.predict([[30, 000, 4, 8]])
```

```python
model.score(X_test,y_test)
```
